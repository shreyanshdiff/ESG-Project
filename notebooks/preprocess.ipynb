```python
import pandas as pd
import numpy as np 
df = pd.read_csv('SP_ESG.csv')
```

```python
df.head()
```

```python
df['Full Time Employees'] = pd.to_numeric(df['Full Time Employees'].str.replace(',',''), errors='coerce')
# Then fill NaN values and convert to int
df['Full Time Employees'] = df['Full Time Employees'].fillna(0).astype('int')
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
numerical_cols = ['Full Time Employees' , 'Total ESG Risk score' , 'Environment Risk Score','Governance Risk Score','Social Risk Score','Controversy Score']
numerical_imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = numerical_imputer.fit_transform(df[numerical_cols])
```

```python
df.isna().sum()
```

```python
# Drop the specified columns
df = df.drop(['Symbol' , 'Name', 'Address', 'Industry', 'Description'], axis=1)
```

```python
df.head()
```

```python
df['Sector'].unique()
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sector'] = le.fit_transform(df['Sector'])
df['Sector'].unique()
```

```python
df['Controversy Level'].nunique()
categorical_cols = ['Controversy Level' , 'ESG Risk Percentile' , 'ESG Risk Level']
```

```python
categorical_imputer = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
df.isna().sum()
```

```python
df.dtypes
```

```python
categorical = ['Controversy Level','ESG Risk Percentile','ESG Risk Level']
for col in categorical:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.to_csv("final.csv")
```
